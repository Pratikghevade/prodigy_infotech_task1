```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import os
```

```python
data=pd.read_csv("IMDB-Movie-Data.csv.xls")
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
Title = df['Title'].head(12)
Rating = df['Rating'].head(12)
```

```python
fig = plt.figure(figsize=(10,7))
```

```python
# Horizontal bar plot
plt.barh(Title[0:10],Rating[0:10])
```

```python
movies_ratings = data[data['Rating'] >= 8.5]
movies_ratings
```

```python
highest_rated_movies = movies_ratings['Rating']
```

```python
# Plotting histogram
plt.hist(highest_rated_movies, bins=6, color='skyblue', edgecolor='black')
# Adding labels and title
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Highest movies rated by IMdB')
```

```python
movies_2015 = data[data['Year'] == 2015]
movies_2015
```

```python
Title = movies_2015['Title'].head(15)
Revenue = movies_2015['Revenue (Millions)'].head(15)
plt.bar(Title, Revenue)
plt.title('Revenue of few movies released in 2015')
plt.xlabel('Title')
plt.ylabel('Revenue (Millions)')
plt.xticks(rotation=90)
plt.show()
```
